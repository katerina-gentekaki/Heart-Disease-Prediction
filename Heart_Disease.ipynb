Heart Disease Prediction Project 
Columns:
1) age: age in years
2) sex: biological sex of the patient:
1 = male
0 = female
3) cp: chest pain type (4 values)
Value 0: typical angina
Value 1: atypical angina
Value 2: non-anginal pain
Value 3: asymptomatic
4) trestbps: resting blood pressure (in mm Hg on admission to the hospital)
5) chol: serum cholestoral in mg/dl
6) fbs: (fasting blood sugar > 120 mg/dl)
1 = true
0 = false
7) restecg: resting electrocardiographic results
Value 0: normal
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
8) thalach: maximum heart rate achieved
exang: exercise induced angina
1 = yes
0 = no
9) Exercise-Induced Angina – Presence of angina induced by exercise:
1 = Yes
0 = No
10) oldpeak = ST depression induced by exercise relative to rest
slope: the slope of the peak exercise ST segment
Value 0: upsloping
Value 1: flat
Value 2: downsloping
11) Slope – The slope of the peak exercise ST segment.
12) ca: number of major vessels (0-3) colored by flourosopy
0 = error (in the original dataset 0 maps to NaN's)
1 = fixed defect
2 = normal
3 = reversable defect
13) thal = Thalassemia– Blood disorder measurement:
3 = Normal
6 = Fixed defect
7 = Reversible defect
14) target (the lable):
0 = no disease,
1 = disease

Splitting the Features and Target

```python
X = df.drop(columns = 'target', axis =1)
Y = df['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Data in Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify= Y, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
#Logistic Regression
model = LogisticRegression()
```

```python
# training the Logistic Regression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy Score on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
#Accuracy Score on Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```

```python
#R² score -R² is not reliable in this case because the dataset is very small
from sklearn.metrics import r2_score
Y_pred = model.predict(X_test)
model_r2_score = round(r2_score(Y_test, Y_pred) * 100, 2)

print("R2 Score for predicted value:", model_r2_score, "%")
```

```python
#F1 score
from sklearn.metrics import f1_score
test_data_f1 = f1_score(Y_test, X_test_prediction)
print('F1 Score on Test data:', round(test_data_f1 * 100, 2), '%')
```

Second Model: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()

# Create and train Decision Tree model
dt_model.fit(X_train, Y_train)
# Predict on test data
dt_test_prediction = dt_model.predict(X_test)
```

```python
# Accuracy of Decision Tree model
dt_test_accuracy = accuracy_score(Y_test, dt_test_prediction)
print('Decision Tree Accuracy on Test data:', round(dt_test_accuracy * 100, 2), '%')
```

```python
# F1 Score of Decision Tree model
dt_test_f1 = f1_score(Y_test, dt_test_prediction)
print('Decision Tree F1 Score on Test data:', round(dt_test_f1 * 100, 2), '%')
```

Compare both models

```python
print(f"Logistic Regression - Accuracy: {round(test_data_accuracy*100,2)}%, F1 Score: {round(test_data_f1*100,2)}%")
print(f"Decision Tree      - Accuracy: {round(dt_test_accuracy*100,2)}%, F1 Score: {round(dt_test_f1*100,2)}%")
```
